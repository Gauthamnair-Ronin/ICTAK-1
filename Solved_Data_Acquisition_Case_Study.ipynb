## Q13. Identify any potential outliers in a numerical column of your choice. Explain your approach.


```python
'''Approach - First we calcualate the queartiles. Then we get Inter Quartile Range from the Q1 and Q3. Ince we get IQR, we use it to define lower bound and upper bound. Data that lies beyond these limits are called outliers'''
# CALCULATING IQR
Q1 = df['windspeed'].quantile(0.25)
Q3 = df['windspeed'].quantile(0.75)

IQR = Q3 - Q1

# OUTLIER VALUES
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['windspeed'] < lower_bound) | (df['windspeed'] > upper_bound)]
outliers
```

## Q.15 Based on your analysis, provide a brief summary of any insights or patterns you discovered in the dataset.

```python
# Write your code here
# extracting weather impact on bike rental

# Correlation matrix for weather variables
weather_corr = df[['cnt', 'temp', 'atemp', 'hum', 'windspeed', 'weathersit']].corr()

print(weather_corr)

# Visualize correlation
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(weather_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Weather Variables and Bike Rentals")
plt.show()
```

```python
# Scatter plot for temperature vs. bike rentals
sns.scatterplot(data=df, x='temp', y='cnt', hue='season', palette="coolwarm")
plt.title("Bike Rentals vs Temperature")
plt.xlabel("Temperature")
plt.ylabel("Bike Rentals (cnt)")
plt.show()

# Scatter plot for humidity vs. bike rentals
sns.scatterplot(data=df, x='hum', y='cnt', hue='season', palette="coolwarm")
plt.title("Bike Rentals vs Humidity")
plt.xlabel("Humidity")
plt.ylabel("Bike Rentals (cnt)")
plt.show()

# Scatter plot for windspeed vs. bike rentals
sns.scatterplot(data=df, x='windspeed', y='cnt', hue='season', palette="coolwarm")
plt.title("Bike Rentals vs Windspeed")
plt.xlabel("Windspeed")
plt.ylabel("Bike Rentals (cnt)")
plt.show()
```

# Analysis
 - The scatter plot *Bike Rental vs Temperature* reveals a positive correlation between temperature and bike rentals, with higher rentals observed at warmer temperatures, especially between 0.4 and 0.8 on the temperature scale. Seasonal patterns are evident, as Fall (Season 4) and Summer (Season 3) see the highest rentals, while Winter (Season 1) has lower activity across all temperatures.

 - The scatter plot 2 i.e *Bike Rental vs Humidity* shows that bike rentals decrease as humidity increases, with the highest rentals occurring at moderate humidity levels (0.4–0.6). 

 - The scatter plot shows that bike rentals tend to decrease as windspeed increases, with the highest concentration of rentals occurring at lower windspeed values (below 0.2). Seasonal patterns reveal that Fall (Season 4) and Summer (Season 3) consistently have higher rentals across all windspeed levels compared to Winter (Season 1).

## Q.16 In which season (Spring, Summer, Fall, Winter) people rented bikes the most?

```python
# Write your code here

season_rentals = df.groupby('season')['cnt'].sum()

#season with the maximum rentals
max_season = season_rentals.idxmax()
max_rentals = season_rentals.max()

season_mapping = {
    1: 'Winter',
    2: 'Spring',
    3: 'Summer',
    4: 'Fall'
}

print("Total bike rentals by season:")
print(season_rentals)

print(f"\n {season_mapping[max_season]} Season had the highest bike rentals with a total of {max_rentals} rentals.")
```

## Q.17 What is the peak hour in which bike rents the most?

```python
# Write your code here
hourly_rentals = df.groupby('hr')['cnt'].sum()


peak_hour = hourly_rentals.idxmax()  
peak_rentals = hourly_rentals.max()  

print("Total bike rentals by hour:")
print(hourly_rentals)

print(f"\nThe peak hour for bike rentals is {peak_hour}:00 with {peak_rentals} rentals.")
```

## Q.18 In which day of a week bikes rents out most?

```python
# Write your code here

weekday_mapping = {
    0: 'Sunday',
    1: 'Monday',
    2: 'Tuesday',
    3: 'Wednesday',
    4: 'Thursday',
    5: 'Friday',
    6: 'Saturday'
}

weekday_rentals = df.groupby('weekday')['cnt'].sum()

max_weekday = weekday_rentals.idxmax()  
max_rentals = weekday_rentals.max()  

max_weekday_name = weekday_mapping[max_weekday]

print("Total bike rentals by day of the week:")
print(weekday_rentals)

print(f"\nThe day with the most bike rentals is {max_weekday_name} with {max_rentals} rentals.")
```
