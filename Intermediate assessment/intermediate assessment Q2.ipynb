```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
df= pd.read_csv("diabetes_model.csv")
df
```

```python
number_of_colors = 8

hexadecimal_alphabets = '0123456789ABCDEF'

color = ["#" + ''.join([random.choice(hexadecimal_alphabets) for j in
range(6)]) for i in range(number_of_colors)]
for i,j in zip(df.columns, range(0,8)):
    lst=df[i].tolist()
    plt.hist(lst,color=color[j],rwidth=0.8,density=True,cumulative=True)
    plt.title(f"Hsitogram of {i} ")
    plt.xlabel(i)
    #plt.ylabel("Density")
    plt.show()
```
