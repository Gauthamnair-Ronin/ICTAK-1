```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("ny-flights.csv")
df
```

### Q1) How much data is available ?

```python
print(f" There are a total of {df.shape[0]} rows and {df.shape[1]} columns of data about the flight details of year 2014")
```

### Q2) How many airlines got cancelled in the year 2014 in total?

```python
print('Total number of cancelled flights in the year 2014 is ',df['cancelled'][df['cancelled']>0].count())
```

### Q3) Which flight had the most delayed arrival time in the year 2014?

```python
sorted_df=df.sort_values(by='arr_delay',ascending =False)
delayed_flight = sorted_df.iloc[0]
print(f" The most late arrived flight in yr 2014 is flight \n#{delayed_flight['fl_num']} that travelled from {delayed_flight['origin']} to {delayed_flight['dest']} on {delayed_flight['dep']}")
```

### Q4) Total number of flight data with missing values 

```python
# separating numerical and non numerial columns

num_df= df[df.columns[df.dtypes!='object']]
obj_df= df[df.columns[df.dtypes=='object']]
```

```python
print('Total number of missing data in numerical  data is\n',num_df.isnull().sum())
```

```python
print('Total number of missing data in non numerical  data is\n',obj_df.isnull().sum())
```

### Q5) Replace numerical missing data with 0 and non numerical with "data not available"

```python
num_df=num_df.fillna(0)
print('Total number of missing data in numerical  data is\n',num_df.isnull().sum())
```

```python
obj_df=obj_df.fillna('Data Not Available')
print('Total number of missing data in non numerical  data is\n',obj_df.isnull().sum())
```

```python
df2=pd.concat([obj_df,num_df],axis=1)
df2
```

### Q6) Which travel plan has experienced the most number or delays in the year 2014?

```python
org_dest_df= df2.groupby(['origin','dest']).max()
org_dest_df=org_dest_df.sort_values(by='arr_delay', ascending=False)
print(f"The Flight from {org_dest_df.index[0][0]} to {org_dest_df.index[0][1]} has experienced the maximum Arrival delay in year 2014")
#print(org_dest_df[['arr_delay','dep_delay']])
org_dest_df=org_dest_df.sort_values(by='dep_delay',ascending=False)
print(f"The Flight from {org_dest_df.index[0][0]} to {org_dest_df.index[0][1]} has experienced the maximum Departure delay in year 2014")
#print(org_dest_df[['arr_delay','dep_delay']])
```

###  Q7) Calculate the longest flight of year 2014

```python
df2
```

```python
# calculating flight duration
def flight_duration(row):
    if (row['arr']=='Data Not Available' or row['dep']=='Data Not Available'):
        return -1000
    else:
        return max(row['arr_time'],row['dep_time'])-min(row['arr_time'],row['dep_time'])

df2['Flight Duration']= df2.apply(flight_duration,axis=1)
df2
```

```python
df2=df2.sort_values(by='Flight Duration',ascending=False)
df2
```

```python
print(f"The flight that took the maximum duration in yr 2014 was \n Flight num    :{df2.iloc[0]['fl_num']}\n from    {df2.iloc[0]['origin']} to {df2.iloc[0]['dest']}\n Duration    :{df2.iloc[0]['Flight Duration']} minutes\n Arrival Date :{df2.iloc[0]['arr']}")
```
